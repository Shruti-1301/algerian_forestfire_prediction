```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline 
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header = 1)
dataset.head()
```

```python
dataset.info()
```

```python
# data cleaning
```

```python
## missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset convert into 2 sets 0 for bejai and 1 ofr sidi-bel
```

```python
dataset.loc[:122,"region"]=0
dataset.loc[122:,"region"]=1
df=dataset
df.head()
df.iloc[122]
```

```python
df[['region']] = df[['region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#removing 122 row null value
df = df.dropna().reset_index(drop=True)
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
df.columns
```

```python
## fix spaces in col name
df.columns=df.columns.str.strip()
df.columns
```

```python
## change datatypes of columns
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)

df.info()
```

```python
##changing other col to float datatype
##generic way(list comprehension)
objects=[features for features in df.columns if df[features].dtypes=='O']

for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## save the cleaned data
df.to_csv('Algerian_forest_fires_cleaned_data.csv',index=False)
```
